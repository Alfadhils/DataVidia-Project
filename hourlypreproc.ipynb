```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
from datetime import datetime
from collections import Counter
import math
```

```python
path = 'dataset/train_hourly.csv'
df = pd.read_csv(path)
df.dropna(axis=0, inplace=True)
df = df.reset_index().drop(columns='index')
df.head()
```

```python
df['time'] = df['time'].astype('string').str[:10]
```

```python
df.groupby(by=['city','time'])
```

```python
uniqcity = Counter(df.city)
dfpercity = []
for city in uniqcity :
    index_range = df.loc[df['city'] == city].index
    dfpercity.append(df[min(index_range):max(index_range)].groupby(by=['time','city']).mean())
            
```

```python
dfnew = pd.concat(dfpercity)
```
